```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data Collection & Pre-Processing**

```python
# breast_cancer_dataset= sklearn.datasets.load_breast_cancer()
# breast_cancer_dataset



# #loading the data into frame
# df= pd.DataFrame(breast_cancer_dataset, columns=breast_cancer_dataset.feature_names)
# df.head()
```

```python

df= pd.read_csv('1st.csv')
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
# Checking missing value
df.isnull().sum()
```

```python
df.describe()
```

```python
df = df.drop(columns=['Unnamed: 32','id'])  # Remove empty column
```

```python
df.info()
```

```python
df['diagnosis'].value_counts()
```

```python
df['diagnosis'] = df['diagnosis'].replace({'M': 0, 'B': 1})
```

```python
df['diagnosis'].value_counts()
```

```python
df.groupby('diagnosis').mean()
```

```python
#Spleatting the features

X = df.drop(columns='diagnosis', axis=1)
Y = df['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Training & Testing Data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)  # 20% tset And 80% training Data
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
print(X_train.shape, Y_train.shape)
```

```python
# training the Logistic Regression model using Training data

model.fit(X_train, Y_train)
```

**Accuracey**

```python
# accurecy check on training data
X_train_prediction = model.predict(X_train)
training_data_acc = accuracy_score(Y_train,X_train_prediction)
```

```python
training_data_acc
```

```python
# accurecy check on test data
X_test_prediction = model.predict(X_test)
test_data_acc = accuracy_score(Y_test,X_test_prediction)
```

```python
test_data_acc
```
